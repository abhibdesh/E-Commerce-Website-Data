# Set Y as Yearly Amount Spent and X as other parameters

```python
X=customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
y=customers['Yearly Amount Spent']
```

# Splitting the dataset in train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

# Importing LinearRegression from sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression
```

# Creating an instance for LinearRegression()

```python
lm=LinearRegression()
```

# Fit/Train the data

```python
lm.fit(X_train,y_train)
```

# Printing the coefficients of the model

```python
lm.coef_
```

# Using predict() to predict x_test set

```python
a=lm.predict(X_test)
```

# Creating a scatterplot of Real values and Predicted values

```python
a=lm.predict(X)
b=customers['Yearly Amount Spent']
plt.figure(figsize=(13,7))
plt.scatter(a,b)
plt.show()
```

# Histogram of residual of the values

```python
sns.displot((a),bins=50)
```

# Checking Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy = lm.score(X_test,y_test)
accuracy
```
